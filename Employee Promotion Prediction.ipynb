```python
# function for replacing null with mean values of previous year rating

def removenull(train):
    train['previous_year_rating']= train['previous_year_rating'].fillna(0)
    train['previous_year_rating']= train['previous_year_rating'].replace(0,train['previous_year_rating'].mean())
    
    
removenull(train)    
    
```
